# 05 - Optional OpenRouter Smoke Test

This notebook is optional. It makes a real OpenRouter API call only if `OPENROUTER_API_KEY` is configured.

**Warning:** Running this notebook may use free-model quota or API credits. It never prints your API key.

```python
import os
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.agent.config import load_config
from src.agent.llm_client import LLMClientError, OpenRouterClient
```

## Check Configuration

This checks whether the API key exists. It does not print the key.

```python
config = load_config()
print("OpenRouter base URL:", config.openrouter_base_url)
print("OpenRouter model:", config.openrouter_model)
print("API key configured:", bool(config.openrouter_api_key))
```

## Tiny Live Call

This cell calls OpenRouter only when an API key exists.

```python
if not config.openrouter_api_key:
    print("Skipping live call because OPENROUTER_API_KEY is not configured.")
else:
    client = OpenRouterClient(config)
    try:
        response = client.complete(
            [{"role": "user", "content": "Reply with exactly: OK"}],
            max_tokens=10,
        )
        print("OpenRouter response:", response)
    except LLMClientError as exc:
        print("OpenRouter call failed:", exc)
```
